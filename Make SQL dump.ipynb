```python
import sqlite3

def view_students():
    # Koneksi ke database
    conn = sqlite3.connect('users.db')
    cur = conn.cursor()

    # Menjalankan query untuk melihat data siswa
    cur.execute("SELECT * FROM students")
    rows = cur.fetchall()

    # Menampilkan data
    for row in rows:
        print(row)

    # Menutup koneksi
    conn.close()

# Panggil fungsi untuk melihat data siswa
view_students()
```
